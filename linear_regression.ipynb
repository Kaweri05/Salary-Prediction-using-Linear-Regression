```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv')
df.head()
```

```python
df.isnull()
```

```python
df.duplicated().sum()
```

```python
df.describe( )
```

```python
x=np.array(df['Experience Years'])
y=np.array(df['Salary'])
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=df[['Experience Years']]
y=df[['Salary']]
plt.scatter(x,y)
plt.figure(figsize=(10,3))
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.show()
```

```python
df.describe().sum()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[10]])
```

```python
year=int(input("enter your experience in year:"))
salary=lr.predict([[year]])
print(salary)
```
